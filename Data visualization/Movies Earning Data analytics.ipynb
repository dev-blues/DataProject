```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns
```

```python
df=pd.read_csv('movies.csv')
```

```python
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
dff=df.dropna()
```

```python
dff.isnull().sum()
```

```python
#changing the datatypes of the columns

dff['budget']=dff['budget'].astype('int64')
dff['gross']=dff['gross'].astype('int64')
```

```python
dff.info()
```

```python
dff.head(5)
```

```python
#Splittng released released column into release_date & release_country

dff[['released_date','released_country']]=dff['released'].str.split('(',expand=True)
```

```python
dff.info()
```

```python
# Deleting released column

dff.drop('released',axis=1, inplace=True)
```

```python
pd.set_option('display.max_rows',None)
```

```python
#Checking highly grossing movies

dff.sort_values(by='gross',inplace=True,ascending=False)
dff
```

```python
# Scatter plot with budget vs gross
style.use('ggplot')
plt.figure(figsize=(10,5))

plt.scatter(x=dff['budget'], y=dff['gross'],cmap='pla51465816')

plt.xlabel('Budget for Film',fontsize=15)
plt.ylabel('Gross Earnings',fontsize=15)
plt.title('Budget vs Gross',fontsize=20)

plt.show()
```

```python
#Plot budget vs gross using seaborn

plt.figure(figsize=(10,5))

sns.regplot(x=dff['budget'], y=dff['gross'],
       scatter_kws={'color':'#00C957'},
       line_kws={'color':'#104E8B'})

plt.xlabel('Budget for Film',fontsize=15)
plt.ylabel('Gross Earnings',fontsize=15)
plt.title('Budget vs Gross',fontsize=20)

plt.show()
```

```python
dff.corr(method='pearson') #pearson, kendall, spearson
```

```python
#High correlation b/w budget and gross
```

```python
correlation_matrix=dff.corr(method='pearson')
plt.figure(figsize=(10,5))

sns.heatmap(correlation_matrix,annot=True,)

plt.xlabel('Movie Features',fontsize=10)
plt.ylabel('Movie Features',fontsize=10)
plt.title('Correlation Matric for Numeric Features',fontsize=10)
plt.show()
```

```python
#changing Company datatype to random numeric value 
df_numerise=dff

for col_name in df_numerise.columns:
    if(df_numerise[col_name].dtype=='object'):
        df_numerise[col_name]=df_numerise[col_name].astype('category')
        df_numerise[col_name]=df_numerise[col_name].cat.codes #to give numeric value
        
df_numerise.head()
```

```python
df.head()
```

```python
correlation_matrix=df_numerise.corr(method='pearson')
plt.figure(figsize=(16,9))

sns.heatmap(correlation_matrix,annot=True,)

plt.xlabel('Movie Features',fontsize=10)
plt.ylabel('Movie Features',fontsize=10)
plt.title('Correlation Matric for Numeric Features',fontsize=10)
plt.show()
```
