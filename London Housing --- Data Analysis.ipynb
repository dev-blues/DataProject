# (A) Convert the datatype of date column to date-time format.

```python
df.head()
```

```python
df['date']=pd.to_datetime(df.date)
df.dtypes
```

# (B.1) Add new column 'year' in the dataframe, which contains year only.

```python
df['year']=df.date.dt.year
df
```

# (B.2) Add new column 'month' as 2nd column in the dataframe, which contains month only.

```python
df.insert(2,'month',df.date.dt.month)
```

```python
df
```

# (C.) Remove the column 'year' and 'month' from the dataframe

```python
df.drop(columns=['year','month'])
```

# (E) What is the min. & max. 'average_price'  per year in England?

```python
df.groupby('year').average_price.max()
```

```python
df.groupby('year').average_price.min()
```

# (F) What is the min. & max. No. of Crimes recorded per year area?

```python
df.head(2)
```

```python
df.groupby('area').no_of_crimes.max().sort_values(ascending=False)
```

```python
df.groupby('area').no_of_crimes.min().sort_values(ascending=False)
```
